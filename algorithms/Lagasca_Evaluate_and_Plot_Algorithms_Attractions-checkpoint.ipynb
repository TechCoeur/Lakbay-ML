```python
def plot_cross_val_measures(x,model):
    x = pd.DataFrame(x)
    plt.plot(x.test_mae, 'v--', label='test_mae')
    plt.plot(x.test_rmse, 'o--', label='test_rmse')
    plt.plot(x.test_fcp, 'o--', label='test_fcp')
    
    labels = np.arange(1, 50, 4)
    plt.xticks(labels, horizontalalignment='right')
    
    plt.legend(loc='best')
    plt.suptitle('%s_measures' % model)
    plt.xlabel('Folds')
    plt.savefig("%s_measures.jpg" % model,dpi=300, bbox_inches='tight')
    plt.show()

def plot_cross_val_time(x,model):
    x = pd.DataFrame(x)
    plt.plot(x.test_time, 'o--', label='test_time')
    plt.plot(x.fit_time, 'v--', label='fit_time')
    labels = np.arange(1, 50, 4)
    plt.xticks(labels)#, horizontalalignment='right')
    #plt.xticks(np.arange(1, 5, 1))
    plt.legend(loc='best')
    plt.suptitle('%s_time' % model)
    plt.xlabel('Folds')
    plt.savefig('%s_time.jpg' % model, dpi=300, bbox_inches='tight')
    plt.show()
```
