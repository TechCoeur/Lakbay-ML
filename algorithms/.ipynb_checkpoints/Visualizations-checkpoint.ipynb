```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_model_constraints(dataset,model):
    df = dataset
    x = (df.drop(columns=["model"])- (pd.DataFrame().append([df.mean().to_frame().transpose()]*8,ignore_index=True)))/(pd.DataFrame().append([df.std().to_frame().transpose()]*8,ignore_index=True))
    plt.plot(x["Fitting Time (seconds)"], '^--', label='Fitting Time (seconds)')
    plt.plot(x["increment memory (MiB)"], 'v--', label='increment memory (MiB)')
    plt.plot(x["Inference Time (seconds)"], 'o--', label='Inference Time (seconds)')
    
    labels = ["Baseline ALS", "Baseline SGD", "NMF", "SVD",
              "SVDpp", "SlopeOne", "CoClustering", "NormalPredictor"]
    
    plt.xticks(np.arange(8),labels,rotation=30, horizontalalignment='right')
    plt.legend(loc='best')
    plt.suptitle('%s ' % model)
    plt.xlabel('Algorithm')
    plt.savefig("%s.jpg" % model,dpi=300, bbox_inches='tight')
    plt.show()
```

```python
def details_difference(dataset,model):
    df = dataset
    x = (df.drop(columns=["model"])- (pd.DataFrame().append([df.mean().to_frame().transpose()]*8,ignore_index=True)))/(pd.DataFrame().append([df.std().to_frame().transpose()]*8,ignore_index=True))
    x["diff"] = x["increment memory (MiB)"]-x["Fitting Time (seconds)"]-x["Inference Time (seconds)"]
    print (model)
    display(x["diff"].sort_values())
```

```python
df = pd.read_csv('/home/dragan21/Documents/laksbay/algorithms/Fit comparison hotels.csv')
```

```python
plot_model_constraints(df,"Hotels")
```

```python
details_difference(df,"Hotels")
```

The top three models for the dataset on hotels, that have the least difference on the three aspects, (starting from the least) are SVD, SlopeOne, and Baseline SGD.
